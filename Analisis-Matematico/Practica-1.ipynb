## Ejercicio 1 — Gráficos, crecimiento/decrecimiento, imagen e inyectividad/sobreyectividad

### a) $f(x) = 4x+1$

Función lineal con pendiente $4>0$: **creciente en todo $\mathbb{R}$**. $\mathrm{Im}(f) = \mathbb{R}$.
Es inyectiva (pendiente $\neq 0$) y sobreyectiva ($\mathbb{R}\to\mathbb{R}$) $\Rightarrow$ **biyectiva**.


```python
%%manim -ql --format gif -v WARNING Ej1a

class Ej1a(Scene):
    def construct(self):
        titulo = Text("a) f(x) = 4x + 1", font_size=32).to_edge(UP)
        axes = Axes(x_range=[-4, 4, 1], y_range=[-6, 10, 2],
                     axis_config={"include_numbers": True}).scale(0.85)
        grafico = axes.plot(lambda x: 4*x + 1, color=BLUE, x_range=[-4, 4])
        self.play(Write(titulo))
        self.play(Create(axes))
        self.play(Create(grafico))
        self.wait()
```

### b) $f(x) = -2x+5$

Función lineal con pendiente $-2<0$: **decreciente en todo $\mathbb{R}$**. $\mathrm{Im}(f) = \mathbb{R}$.

```python
%%manim -ql --format gif -v WARNING Ej1b

class Ej1b(Scene):
    def construct(self):
        titulo = Text("b) f(x) = -2x + 5", font_size=32).to_edge(UP)
        axes = Axes(x_range=[-4, 4, 1], y_range=[-6, 14, 2],
                     axis_config={"include_numbers": True}).scale(0.85)
        grafico = axes.plot(lambda x: -2*x + 5, color=RED, x_range=[-4, 4])
        self.play(Write(titulo))
        self.play(Create(axes))
        self.play(Create(grafico))
        self.wait()
```

### c) $f(x) = -2(x-1)^2+3$

**Creciente en $(-\infty,1]$, decreciente en $[1,+\infty)$.** $\mathrm{Im}(f) = (-\infty,3]$.

```python
%%manim -ql --format gif -v WARNING Ej1c

class Ej1c(Scene):
    def construct(self):
        titulo = Text("c) f(x) = -2(x-1)^2 + 3", font_size=32).to_edge(UP)
        axes = Axes(x_range=[-3, 5, 1], y_range=[-15, 5, 2],
                     axis_config={"include_numbers": True}).scale(0.75)
        grafico = axes.plot(lambda x: -2*(x-1)**2 + 3, color=GREEN, x_range=[-2.2, 4.2])
        vertice = Dot(axes.c2p(1, 3), color=YELLOW)
        label = MathTex("(1,\\,3)", font_size=28).next_to(vertice, UP)
        self.play(Write(titulo))
        self.play(Create(axes))
        self.play(Create(grafico))
        self.play(Create(vertice), Write(label))
        self.wait()
```

### e) $f(x)=\begin{cases} x & x\le 0\\ 3x & 0<x<2\\ -x^2+10 & x\ge 2\end{cases}$

La función es continua (empalma en $(0,0)$ y en $(2,6)$).
**Creciente en $(-\infty,2]$, decreciente en $[2,+\infty)$** (máximo en $x=2$, $f(2)=6$).
$\mathrm{Im}(f) = (-\infty,6]$ (el tercer tramo por sí solo ya cubre todo ese rango).
No es inyectiva (p. ej. $f(-1)=-1$ y también $f(\sqrt{11})=-1$, con $\sqrt{11}\ge2$) ni sobreyectiva $\Rightarrow$ **no es biyectiva**.


```python
%%manim -ql --format gif -v WARNING Ej1e

class Ej1e(Scene):
    def construct(self):
        titulo = Text("e) función partida", font_size=32).to_edge(UP)
        axes = Axes(x_range=[-4, 5, 1], y_range=[-4, 8, 2],
                     axis_config={"include_numbers": True}).scale(0.8)
        tramo1 = axes.plot(lambda x: x, color=GREEN, x_range=[-4, 0])
        tramo2 = axes.plot(lambda x: 3*x, color=BLUE, x_range=[0, 2])
        tramo3 = axes.plot(lambda x: -x**2 + 10, color=RED, x_range=[2, 4.3])
        p1 = Dot(axes.c2p(0, 0), color=WHITE)
        p2 = Dot(axes.c2p(2, 6), color=WHITE)
        self.play(Write(titulo))
        self.play(Create(axes))
        self.play(Create(tramo1))
        self.play(Create(tramo2))
        self.play(Create(tramo3))
        self.play(Create(p1), Create(p2))
        self.wait()
```

### f) $f(x)=\begin{cases} x-1 & x\le 2\\ x^2-9 & 2<x<4\\ 3x-5 & x\ge 4\end{cases}$

Hay un **salto de discontinuidad en $x=2$**: $f(2)=1$ pero $\lim_{x\to 2^+}f(x) = -5$. En $x=4$ sí empalma ($4^2-9 = 3\cdot4-5=7$).

Por tramos, $f$ es creciente en $(-\infty,2]$ y creciente en $(2,+\infty)$, pero **no es creciente en todo $\mathbb{R}$** por el salto hacia abajo en $x=2$.

Imagen por tramo: $(-\infty,1]\ \cup\ (-5,7)\ \cup\ [7,+\infty)$. Estos tres conjuntos, unidos, cubren **todo $\mathbb{R}$** $\Rightarrow$ $f$ es **sobreyectiva**.
Sin embargo no es inyectiva (p. ej. $f(1)=0$ y $f(3)=0$) $\Rightarrow$ **no es biyectiva**, a pesar de ser sobreyectiva.


```python
%%manim -ql --format gif -v WARNING Ej1f

class Ej1f(Scene):
    def construct(self):
        titulo = Text("f) función partida", font_size=32).to_edge(UP)
        axes = Axes(x_range=[-2, 7, 1], y_range=[-6, 14, 2],
                     axis_config={"include_numbers": True}).scale(0.75)
        tramo1 = axes.plot(lambda x: x - 1, color=GREEN, x_range=[-2, 2])
        tramo2 = axes.plot(lambda x: x**2 - 9, color=BLUE, x_range=[2, 4])
        tramo3 = axes.plot(lambda x: 3*x - 5, color=RED, x_range=[4, 6.3])
        salto1 = Dot(axes.c2p(2, 1), color=WHITE)
        hueco1 = Dot(axes.c2p(2, -5), color=BLUE, fill_opacity=0).set_stroke(BLUE, width=2)
        empalme = Dot(axes.c2p(4, 7), color=WHITE)
        self.play(Write(titulo))
        self.play(Create(axes))
        self.play(Create(tramo1))
        self.play(Create(tramo2))
        self.play(Create(tramo3))
        self.play(Create(salto1), Create(hueco1), Create(empalme))
        self.wait()
```

- **f)** Sólo el tercer tramo: $3x-5=7 \Rightarrow x=4$, y $4\ge4$ es válido. Conjunto: $\{4\}$.


## Ejercicio 3 — Dominio natural e intersecciones con los ejes



### a) $f(x) = \sqrt{3x-1}$

Dominio: $3x-1\ge 0 \Rightarrow x\ge \tfrac13$, o sea $\mathrm{Dom}(f)=\left[\tfrac13,\ +\infty\right)$.

- Intersección con el eje $x$ (resolver $f(x)=0$): $3x-1=0 \Rightarrow x=\tfrac13$. Punto $\left(\tfrac13,0\right)$.
- Intersección con el eje $y$ (evaluar $f(0)$): $f(0)=\sqrt{-1}$, no existe ($0\notin \mathrm{Dom}(f)$) → no hay intersección con el eje $y$.

*(Nota: en el intento anterior estas dos intersecciones estaban con las etiquetas cruzadas — acá quedan corregidas.)*

### b) $f(x) = \dfrac{x^2+1}{x+2}$

Dominio: $x\neq -2 \Rightarrow \mathrm{Dom}(f) = \mathbb{R}\setminus\{-2\}$.

- Eje $x$: $x^2+1=0$ no tiene soluciones reales → no hay intersección con el eje $x$.
- Eje $y$: $f(0) = \dfrac{1}{2}$. Punto $\left(0,\tfrac12\right)$.

### c) $f(x) = 2 - \dfrac{x}{\sqrt{x}}$

Como $\dfrac{x}{\sqrt{x}}$ requiere $x>0$ (en $x=0$ queda $0/0$, indefinido), el dominio es $\mathrm{Dom}(f) = (0,+\infty)$ — abierto, **no** incluye a $0$.

- Eje $y$: $0\notin \mathrm{Dom}(f)$ → no hay intersección con el eje $y$.
- Eje $x$: para $x>0$, $\dfrac{x}{\sqrt x}=\sqrt x$, entonces $2-\sqrt x = 0 \Rightarrow x=4$. Punto $(4,0)$.

### d) $f(x) = -x + \sqrt{x+2}$

Dominio: $x+2\ge 0 \Rightarrow \mathrm{Dom}(f) = [-2,+\infty)$.

- Eje $y$: $f(0) = 0+\sqrt{2} = \sqrt2$. Punto $(0,\sqrt2)$.
- Eje $x$: $-x+\sqrt{x+2}=0 \Rightarrow \sqrt{x+2}=x$ (requiere $x\ge 0$) $\Rightarrow x+2=x^2 \Rightarrow x^2-x-2=0 \Rightarrow (x-2)(x+1)=0 \Rightarrow x=2$ o $x=-1$. Como necesitamos $x\ge 0$, se descarta $x=-1$ (raíz espuria introducida al elevar al cuadrado). Punto $(2,0)$.

### e) $f(x) = \dfrac{x^3-1}{\sqrt{x^2-x-2}}$

El radicando debe ser **estrictamente positivo** (está en un denominador): $x^2-x-2=(x-2)(x+1)>0 \Rightarrow x<-1 \text{ o } x>2$.
$\mathrm{Dom}(f) = (-\infty,-1)\cup(2,+\infty)$.

- Eje $y$: $0\notin\mathrm{Dom}(f)$ → no hay intersección con el eje $y$.
- Eje $x$: el numerador se anula en $x^3=1\Rightarrow x=1$, pero $1\notin\mathrm{Dom}(f)$ → no hay intersección con el eje $x$.

### f) $f(x) = \dfrac{x}{1+x^3}$

Dominio: $1+x^3\neq 0 \Rightarrow x\neq -1 \Rightarrow \mathrm{Dom}(f)=\mathbb{R}\setminus\{-1\}$.

- Eje $y$ y eje $x$ coinciden: $f(0)=0$. Único punto de intersección: $(0,0)$.

## Ejercicio 5 — Composición de funciones

$f(x)=2x^2+5x,\qquad g(x)=\dfrac{1}{x+3},\qquad h(x)=2x-6$

### a) Valores puntuales

- $(f\circ f)(-1) = f(f(-1)) = f(-3) = 3$
- $(f\circ h)(1) = f(h(1)) = f(-4) = 12$
- $(g\circ f)(-1) = g(f(-1)) = g(-3)$: como $g(x)=\tfrac{1}{x+3}$, $g(-3)$ no existe (división por cero) → **no está definido**.
- $(h\circ g)(2) = h(g(2)) = h\!\left(\tfrac15\right) = -\tfrac{28}{5}$

### b) Fórmulas

- $(f\circ g)(x) = \dfrac{5x+17}{(x+3)^2} = \dfrac{5x+17}{x^2+6x+9}$
- $(g\circ f)(x) = \dfrac{1}{2x^2+5x+3}$
- $((f\circ g)\circ h)(x) = \dfrac{10x-13}{(2x-3)^2} = \dfrac{10x-13}{4x^2-12x+9}$
- $(f\circ h)(x) = 8x^2-38x+42$
- $(f\circ f)(x) = 8x^4+40x^3+60x^2+25x$

### c) ¿Son $f\circ g$ y $g\circ f$ la misma función?

No. Alcanza con comparar sus dominios/expresiones: $f\circ g$ tiene como restricción $x\neq -3$ y es una función racional con denominador cuadrático $(x+3)^2$; $g\circ f$ excluye los valores de $x$ donde $2x^2+5x=-3$ y tiene una expresión totalmente distinta. Además, evaluando por ejemplo en $x=0$: $(f\circ g)(0)=\tfrac{17}{9}$ mientras que $(g\circ f)(0)=\tfrac13$. Son distintas.

## Ejercicio 7 — Polinomios

$p(x)=x^3-x^2+x+3,\qquad q(x)=x-1,\qquad r(x)=2x^2+5$

### a)
- **i)** $p(-2) = -11$
- **ii)** $2p(x)+q(x) = 2x^3-2x^2+3x+5$
- **iii)** $q(x)r(x) = 2x^3-2x^2+5x-5$
- **iv)** $p(x)\div r(x)$: cociente $\dfrac{x}{2}-\dfrac12$, resto $\dfrac{11-3x}{2}$
- **v)** $r(x)\div q(x)$: cociente $2x+2$, resto $7$

### b)
Para que $p(x)+t(x)$ tenga grado 2, $t(x)$ debe cancelar el término $x^3$ de $p$. Alcanza con $t(x)=-x^3$ (que sigue siendo de grado 3):
$$p(x)+t(x) = -x^2+x+3 \quad(\text{grado } 2)$$
