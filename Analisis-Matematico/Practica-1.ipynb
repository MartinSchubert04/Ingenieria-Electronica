```python
%%manim -qm -v WARNING GraficoFunciones

class GraficoFunciones(Scene):
    def construct(self):
        # ---------- a) f(x) = 4x + 1 (lineal) ----------
        titulo = Text("a) f(x) = 4x + 1", font_size=36).to_edge(UP)
        axes1 = Axes(
            x_range=[-3, 3, 1], y_range=[-5, 10, 2],
            axis_config={"include_numbers": True}
        )
        grafico1 = axes1.plot(lambda x: 4*x + 1, color=BLUE, x_range=[-3, 3])

        self.play(Write(titulo))
        self.play(Create(axes1))
        self.play(Create(grafico1))
        self.wait(1)

        # Nota de análisis (creciente/decreciente, imagen, tipo)
        nota1 = Text(
            "Creciente en todo su dominio\nImagen: R\nBiyectiva",
            font_size=24, color=YELLOW
        ).to_corner(DR)
        self.play(Write(nota1))
        self.wait(2)

        self.play(FadeOut(titulo, axes1, grafico1, nota1))
        
        # ---------- b) f(x) = -2(x-1)^2 + 3 (cuadrática) ----------
        titulo2 = Text("f(x) = -2x + 5", font_size=36).to_edge(UP)
        axes2 = Axes(
            x_range=[-3, 3, 1], y_range=[-5, 10, 2],
            axis_config={"include_numbers": True}
        )
        grafico2= axes2.plot(lambda x: -2*x + 5, color=RED)
        self.play(Write(titulo2))
        self.play(Create(axes2))
        self.play(Create(grafico2))
        self.wait(1)
        
        nota2 = Text(
            "Decreciente en todo su dominio\nImagen: R\nBiyectiva",
            font_size=24, color=YELLOW
        ).to_corner(DR)
        self.play(Write(nota2))
        self.wait(2)
        
        self.play(FadeOut(titulo2, axes2, grafico2, nota2))
        
        # ---------- c) f(x) = -2(x-1)^2 + 3 (cuadrática) ----------
        titulo3 = Text("c) f(x) = -2(x-1)² + 3", font_size=36).to_edge(UP)
        axes3 = Axes(
            x_range=[-3, 5, 1], y_range=[-10, 5, 2],
            axis_config={"include_numbers": True}
        )
        grafico3 = axes2.plot(lambda x: (-2*(x-1)**2) + 3, color=RED, x_range=[-2, 4])

        # Marcar el vértice
        vertice = Dot(axes3.c2p(1, 3), color=YELLOW)
        label_vertice = MathTex("(1, 3)", font_size=28).next_to(vertice, UP)

        self.play(Write(titulo3))
        self.play(Create(axes3))
        self.play(Create(grafico3))
        self.play(Create(vertice), Write(label_vertice))
        self.wait(1)

        nota3 = Text(
            "Creciente en (-∞, 1)\nDecreciente en (1, +∞)\n"
            "Imagen: (-∞, 3]\nNi inyectiva ni sobreyectiva en R→R",
            font_size=22, color=YELLOW
        ).to_corner(DR)
        self.play(Write(nota3))
        self.wait(2)

        self.play(FadeOut(titulo3, axes3, grafico3, vertice, label_vertice, nota3))



        # ---------- e) función piecewise ----------
        titulo4 = Text("e) Función partida", font_size=36).to_edge(UP)
        axes4 = Axes(
            x_range=[-3, 5, 1], y_range=[-5, 12, 2],
            axis_config={"include_numbers": True}
        )

        # Cada tramo se grafica por separado, respetando su dominio
        tramo1 = axes4.plot(lambda x: x, color=GREEN, x_range=[-3, 0])
        tramo2 = axes4.plot(lambda x: 3*x, color=BLUE, x_range=[0, 2])
        tramo3 = axes4.plot(lambda x: -x**2 + 10, color=RED, x_range=[2, 5])

        self.play(Write(titulo4))
        self.play(Create(axes4))
        self.play(Create(tramo1))
        self.play(Create(tramo2))
        self.play(Create(tramo3))
        self.wait(2)
```
