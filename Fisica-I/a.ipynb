## Problema: Tiro oblicuo
Un proyectil se lanza a 30 m/s con ángulo de 45°...

### Datos
$v_0 = 30 m/s$, $\theta = 45°$

### Desarrollo
$v_x = v_0 \cos\theta$

## Animación con Manim

Visualizamos la trayectoria del tiro oblicuo ($v_0 = 30\ m/s$, $\theta = 45°$) usando [Manim](https://docs.manim.community/). Requiere tener instalado `manim` (`pip install manim`) y `ffmpeg` en el sistema.

```python
%pip install manim -q
from manim import *
import numpy as np
```

```python
%%manim -qm -v WARNING TiroOblicuo

class TiroOblicuo(Scene):
    def construct(self):
        v0 = 30
        theta = 45 * DEGREES
        g = 9.8

        vx = v0 * np.cos(theta)
        vy0 = v0 * np.sin(theta)
        t_vuelo = 2 * vy0 / g
        x_max = vx * t_vuelo
        y_max = (vy0 ** 2) / (2 * g)

        axes = Axes(
            x_range=[0, x_max * 1.1, 20],
            y_range=[0, y_max * 1.3, 10],
            axis_config={"include_numbers": True},
        ).add_coordinates()
        axes.to_edge(DOWN)
        labels = axes.get_axis_labels(x_label="x\\,(m)", y_label="y\\,(m)")

        def posicion(t):
            x = vx * t
            y = vy0 * t - 0.5 * g * t ** 2
            return axes.c2p(x, y)

        trayectoria = axes.plot_parametric_curve(
            lambda t: [vx * t, vy0 * t - 0.5 * g * t ** 2, 0],
            t_range=[0, t_vuelo],
            color=YELLOW,
        )

        proyectil = Dot(color=RED).move_to(posicion(0))
        datos = MathTex(
            r"v_0 = 30\ m/s,\ \theta = 45^\circ"
        ).to_edge(UP)

        self.play(Write(datos))
        self.play(Create(axes), Write(labels))
        self.play(Create(trayectoria), run_time=2)
        self.play(MoveAlongPath(proyectil, trayectoria), run_time=3, rate_func=linear)
        self.wait()
```
